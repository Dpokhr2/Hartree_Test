```python
import pandas as pd
```

```python
#Read the first data
df = pd.read_csv('dataset1.csv')
df
```

```python
#Read the second data

df1 = pd.read_csv('dataset2.csv')
df1
```

```python
#Join two data sets
joined_df = pd.merge(df, df1, on='counter_party', how='inner')
joined_df
```

```python
#Find the max rating
max_rating = joined_df.groupby(['counter_party', 'legal_entity', 'tier'])['rating'].max().reset_index()
max_rating = max_rating.rename(columns={'rating': 'max_rating'})
```

```python
max_rating
```

```python
#Sum the value where status = 'ARAP'

filter = joined_df[joined_df['status']=="ARAP"].groupby(['legal_entity', 'counter_party'])['value'].sum().reset_index()
filter
```

```python
#Sum the value where status = 'ACCR'
filter_ACCR = joined_df[joined_df['status']=="ACCR"].groupby(['legal_entity', 'counter_party'])['value'].sum().reset_index()
filter_ACCR
```

```python
output_df = pd.merge(max_rating, filter, on=['legal_entity', 'counter_party'], how='outer')
```

```python
output_df = pd.merge(output_df, filter_ACCR, on=['legal_entity', 'counter_party'], how='outer')
output_df
```

```python
output_df['Total'] = output_df['value_x'] + output_df['value_y']
output_df
```

```python
result_df = output_df[['counter_party', 'legal_entity', 'Total']]
result_df
```

```python
result_df.to_csv('output.csv')
```
